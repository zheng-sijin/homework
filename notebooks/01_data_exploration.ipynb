```python
# -*- coding: utf-8 -*-
"""
01_data_exploration.ipynb - 粤港澳大湾区数据要素流动探索性数据分析

本笔记本完成：
1. 数据加载与合并
2. 描述性统计分析
3. 相关性分析
4. 时间趋势分析
5. 空间自相关分析
6. 高级统计分析（基尼系数、泰尔指数等）

输出保存至：outputs/figures/ 和 outputs/tables/
"""

import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# 添加项目根目录到路径
sys.path.append('..')
sys.path.append('../src')

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] = False

# 设置样式
sns.set_style("whitegrid")
sns.set_palette("husl")

# 创建输出目录
os.makedirs('../outputs/figures', exist_ok=True)
os.makedirs('../outputs/tables', exist_ok=True)

# %% [markdown]
# # 1. 数据加载与预处理

# %%
# 加载主数据
df_main = pd.read_csv('../data/raw/main_data_advanced.csv')
print(f"主数据形状: {df_main.shape}")
print(f"城市数量: {df_main['城市'].nunique()}")
print(f"年份范围: {df_main['年份'].min()} - {df_main['年份'].max()}")

# 加载OD矩阵数据
od_files = {
    2019: '../data/raw/od_matrix_2019.csv',
    2020: '../data/raw/od_matrix_2020.csv', 
    2021: '../data/raw/od_matrix_2021.csv',
    2022: '../data/raw/od_matrix_2022.csv',
    2023: '../data/raw/od_matrix_2023.csv'
}

od_data = {}
for year, filepath in od_files.items():
    od_data[year] = pd.read_csv(filepath, index_col=0)

print("\nOD矩阵已加载:")
for year, df in od_data.items():
    print(f"{year}年: {df.shape}")

# %% [markdown]
# ## 1.1 数据基本信息

# %%
# 检查主数据基本信息
print("=== 主数据基本信息 ===")
print(df_main.info())

# 检查缺失值
missing_values = df_main.isnull().sum()
missing_percent = (missing_values / len(df_main)) * 100
missing_df = pd.DataFrame({
    '缺失数量': missing_values,
    '缺失比例(%)': missing_percent
}).sort_values('缺失数量', ascending=False)

print("\n=== 缺失值统计 ===")
print(missing_df[missing_df['缺失数量'] > 0].head(20))

# %% [markdown]
# ## 1.2 数据预处理

# %%
# 选择关键指标进行分析（根据实验要求） - 修正：使用实际存在的列名
key_metrics = [
    '跨境数据传输总量_TB', 
    '数据中心数量', 
    '互联网国际出口带宽_Gbps',
    'IaaS使用量_万核时', 
    '政务API调用量_亿次',  # 改为实际存在的列名
    '数据交易额_亿元',
    'GDP_亿元', 
    '数字经济核心产业增加值_亿元', 
    '研发经费投入_亿元',
    '5G基站数量', 
    '算力规模_PFLOPS', 
    '物联网连接数_万个'
]

print(f"关键指标数量: {len(key_metrics)}")
print("关键指标列表:")
for i, metric in enumerate(key_metrics):
    print(f"  {i+1}. {metric}")

# 检查所有指标是否都存在
missing_in_data = [metric for metric in key_metrics if metric not in df_main.columns]
if missing_in_data:
    print(f"\n以下指标在数据中不存在: {missing_in_data}")
else:
    print("\n所有关键指标在数据中都存在")

# 创建处理后的数据副本
df_processed = df_main.copy()

# 填充缺失值（使用前向填充） - 由于数据没有缺失值，这一步可以跳过，但保留以防万一
for metric in key_metrics:
    if df_processed[metric].isnull().any():
        df_processed[metric] = df_processed.groupby('城市')[metric].ffill()
        print(f"{metric}: 填充了 {df_processed[metric].isnull().sum()} 个缺失值")

# 检查填充后的缺失值
print("\n填充后缺失值数量:")
missing_after = df_processed[key_metrics].isnull().sum()
print(missing_after[missing_after > 0])

# 如果没有缺失值
if missing_after.sum() == 0:
    print("所有关键指标都没有缺失值")

# %% [markdown]
# # 2. 描述性统计分析

# %%
# 2.1 基本描述性统计
desc_stats = df_processed[key_metrics].describe().T
desc_stats['变异系数'] = desc_stats['std'] / desc_stats['mean']
desc_stats['偏度'] = df_processed[key_metrics].apply(lambda x: stats.skew(x.dropna()))
desc_stats['峰度'] = df_processed[key_metrics].apply(lambda x: stats.kurtosis(x.dropna()))

print("=== 描述性统计 ===")
print(desc_stats.round(3))

# 保存描述性统计表
desc_stats.to_csv('../outputs/tables/descriptive_statistics.csv', encoding='utf-8-sig')

# %%
# 2.2 时间趋势分析
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.flatten()

for idx, metric in enumerate(key_metrics[:12]):
    ax = axes[idx]
    for city in df_processed['城市'].unique()[:5]:  # 只显示前5个城市
        city_data = df_processed[df_processed['城市'] == city]
        ax.plot(city_data['年份'], city_data[metric], label=city, marker='o', linewidth=2)
    
    ax.set_title(f'{metric}时间趋势', fontsize=12, fontweight='bold')
    ax.set_xlabel('年份')
    ax.set_ylabel(metric)
    ax.legend(loc='best', fontsize=8)
    ax.grid(True, alpha=0.3)

plt.suptitle('关键指标时间趋势分析（前5个城市）', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/figures/time_trends.png', dpi=300, bbox_inches='tight')
plt.show()

# %%
# 2.3 正态性检验
normality_tests = []
for metric in key_metrics:
    data = df_processed[metric].dropna()
    if len(data) > 3:
        stat_sw, p_sw = stats.shapiro(data[:5000]) if len(data) > 5000 else stats.shapiro(data)
        stat_ks, p_ks = stats.kstest(data, 'norm', args=(data.mean(), data.std()))
        
        normality_tests.append({
            '指标': metric,
            'Shapiro-Wilk统计量': stat_sw,
            'Shapiro-Wilk p值': p_sw,
            'K-S统计量': stat_ks,
            'K-S p值': p_ks,
            '是否正态(α=0.05)': p_sw > 0.05 and p_ks > 0.05
        })

normality_df = pd.DataFrame(normality_tests)
print("=== 正态性检验结果 ===")
print(normality_df.round(4))

# 保存正态性检验结果
normality_df.to_csv('../outputs/tables/normality_tests.csv', index=False, encoding='utf-8-sig')

# Q-Q图
fig, axes = plt.subplots(3, 4, figsize=(20, 15))
axes = axes.flatten()

for idx, metric in enumerate(key_metrics[:12]):
    ax = axes[idx]
    data = df_processed[metric].dropna()
    stats.probplot(data, dist="norm", plot=ax)
    ax.set_title(f'{metric} Q-Q图', fontsize=12, fontweight='bold')
    ax.grid(True, alpha=0.3)

plt.suptitle('关键指标Q-Q图（正态性检验）', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/figures/qq_plots.png', dpi=300, bbox_inches='tight')
plt.show()

# %% [markdown]
# # 3. 相关性分析

# %%
# 3.1 相关系数矩阵
corr_matrix = df_processed[key_metrics].corr(method='pearson')

plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, mask=mask, annot=True, fmt='.2f', cmap='RdBu_r',
            center=0, square=True, linewidths=0.5, cbar_kws={"shrink": 0.8})
plt.title('Pearson相关系数矩阵热力图', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('../outputs/figures/correlation_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# 保存相关系数矩阵
corr_matrix.to_csv('../outputs/tables/correlation_matrix.csv', encoding='utf-8-sig')

# %%
# 3.2 偏相关分析（控制GDP影响）
from scipy import stats

def partial_correlation(df, x, y, control):
    """计算偏相关系数"""
    # 手动计算残差
    # 获取x对控制变量的回归斜率和截距
    slope_x, intercept_x, _, _, _ = stats.linregress(df[control], df[x])
    # 计算x的预测值
    x_pred = intercept_x + slope_x * df[control]
    # 计算x的残差
    res_x = df[x] - x_pred
    
    # 获取y对控制变量的回归斜率和截距
    slope_y, intercept_y, _, _, _ = stats.linregress(df[control], df[y])
    # 计算y的预测值
    y_pred = intercept_y + slope_y * df[control]
    # 计算y的残差
    res_y = df[y] - y_pred
    
    # 计算残差的相关系数
    return np.corrcoef(res_x, res_y)[0, 1]
# 选择几个关键变量进行偏相关分析
partial_corr_results = []
variables_to_test = ['跨境数据传输总量_TB', 'IaaS使用量_万核时', '政务API调用量_亿次', '数据交易额_亿元']

for var in variables_to_test:
    if var != 'GDP_亿元':
        pc = partial_correlation(df_processed, var, '数字经济核心产业增加值_亿元', 'GDP_亿元')
        partial_corr_results.append({
            '变量': var,
            '控制变量': 'GDP_亿元',
            '目标变量': '数字经济核心产业增加值_亿元',
            '偏相关系数': pc
        })

partial_corr_df = pd.DataFrame(partial_corr_results)
print("=== 偏相关分析结果（控制GDP） ===")
print(partial_corr_df.round(4))

# 保存偏相关分析结果
partial_corr_df.to_csv('../outputs/tables/partial_correlation.csv', index=False, encoding='utf-8-sig')

# %%
# 3.3 距离相关性（非线性关系）
try:
    from dcor import distance_correlation
    
    # 计算距离相关性
    dist_corr_results = []
    for var in variables_to_test:
        if var != 'GDP_亿元':
            dc = distance_correlation(df_processed[var].dropna(), 
                                     df_processed['数字经济核心产业增加值_亿元'].dropna())
            dist_corr_results.append({
                '变量': var,
                '目标变量': '数字经济核心产业增加值_亿元',
                '距离相关性': dc
            })
    
    dist_corr_df = pd.DataFrame(dist_corr_results)
    print("=== 距离相关性分析结果 ===")
    print(dist_corr_df.round(4))
    
    # 保存结果
    dist_corr_df.to_csv('../outputs/tables/distance_correlation.csv', index=False, encoding='utf-8-sig')
    
except ImportError:
    print("dcor库未安装，跳过距离相关性分析")
    # 安装命令: pip install dcor

# %% [markdown]
# # 4. 空间自相关分析

# %%
# 4.1 加载地理数据（这里使用简化的坐标数据）
cities_coords = {
    '香港': [114.17, 22.32],
    '澳门': [113.54, 22.19],
    '广州': [113.27, 23.13],
    '深圳': [114.07, 22.62],
    '珠海': [113.57, 22.27],
    '佛山': [113.12, 23.02],
    '惠州': [114.42, 23.11],
    '东莞': [113.75, 23.05],
    '中山': [113.38, 22.52],
    '江门': [113.08, 22.58],
    '肇庆': [112.47, 23.05]
}

# 创建地理DataFrame
gdf = gpd.GeoDataFrame(
    pd.DataFrame.from_dict(cities_coords, orient='index', columns=['经度', '纬度']),
    geometry=gpd.points_from_xy(pd.DataFrame.from_dict(cities_coords, orient='index', columns=['经度', '纬度'])['经度'],
                               pd.DataFrame.from_dict(cities_coords, orient='index', columns=['经度', '纬度'])['纬度']),
    crs="EPSG:4326"
)

# 手动计算距离矩阵
from scipy.spatial import distance_matrix

# 计算城市间的距离矩阵
coords_array = np.array(list(cities_coords.values()))
dist_matrix = distance_matrix(coords_array, coords_array)

# 创建KNN权重矩阵（K=3）
k = 3
n_cities = len(cities_coords)
knn_weights = np.zeros((n_cities, n_cities))

for i in range(n_cities):
    # 获取第i个城市的K个最近邻（不包括自己）
    indices = np.argsort(dist_matrix[i])[1:k+1]  # 跳过自己（距离为0）
    knn_weights[i, indices] = 1  # 赋予权重1

print("手动创建的KNN权重矩阵形状:", knn_weights.shape)

# 创建自定义的权重对象
class ManualWeights:
    def __init__(self, weights, cities):
        self.weights = weights
        self.cities = cities
        self.n = len(cities)
        
    def __getitem__(self, i):
        return {self.cities[j]: self.weights[i, j] for j in range(self.n) if self.weights[i, j] > 0}

knn = ManualWeights(knn_weights, list(cities_coords.keys()))

# %%
# 4.2 全局Moran's I检验（手动实现）
print("=== 全局Moran's I检验（手动实现） ===")

# 使用2023年数据计算空间自相关
data_2023 = df_processed[df_processed['年份'] == 2023]
data_2023 = data_2023.set_index('城市').reindex(gdf.index)

# 计算跨境数据传输量的Moran's I
y = data_2023['跨境数据传输总量_TB'].fillna(0).values

# 手动计算Moran's I
n = len(y)
y_mean = np.mean(y)

# 计算空间滞后
spatial_lag = np.zeros(n)
for i in range(n):
    neighbors = knn[i]  # 获取邻居
    if neighbors:
        neighbor_vals = [y[list(cities_coords.keys()).index(city)] for city in neighbors.keys()]
        spatial_lag[i] = np.mean(neighbor_vals)

# 计算Moran's I
numerator = np.sum((y - y_mean) * (spatial_lag - y_mean))
denominator = np.sum((y - y_mean) ** 2)
moran_i = (n / np.sum(knn_weights)) * (numerator / denominator)

print(f"Moran's I: {moran_i:.4f}")

# 计算期望值、方差和Z得分（简化版本）
# Moran's I的期望值：E[I] = -1/(n-1)
expected_i = -1/(n-1) if n > 1 else 0

# 计算方差（简化公式）
S0 = np.sum(knn_weights)
S1 = np.sum((knn_weights + knn_weights.T) ** 2) / 2
S2 = np.sum((np.sum(knn_weights, axis=0) + np.sum(knn_weights, axis=1)) ** 2)

# 方差公式（基于随机排列假设的简化版本）
var_i = (n * (n**2 - 3*n + 3) * S1 - n * S2 + 3 * S0**2) / ((n-1)*(n-2)*(n-3)*S0**2) - expected_i**2

# 计算Z得分
z_score = (moran_i - expected_i) / np.sqrt(var_i) if var_i > 0 else 0

# 计算P值（双侧检验）
from scipy import stats as sp_stats
p_value = 2 * (1 - sp_stats.norm.cdf(abs(z_score)))

print(f"期望值E[I]: {expected_i:.4f}")
print(f"方差: {var_i:.6f}")
print(f"Z得分: {z_score:.4f}")
print(f"P值: {p_value:.6f}")
print(f"显著性(α=0.05): {p_value < 0.05}")

# 绘制Moran散点图
fig, ax = plt.subplots(figsize=(10, 8))

# 绘制散点
scatter = ax.scatter(y, spatial_lag, alpha=0.7, s=100, c=y, cmap='viridis')

# 添加城市标签
for i, city in enumerate(data_2023.index):
    ax.annotate(city, (y[i], spatial_lag[i]), fontsize=9, xytext=(5, 5), textcoords='offset points')

# 添加趋势线
z = np.polyfit(y, spatial_lag, 1)
p = np.poly1d(z)
ax.plot(y, p(y), "r--", alpha=0.8, label=f'趋势线 (斜率={z[0]:.3f})')

# 添加象限线
ax.axhline(y=y_mean, color='gray', linestyle='--', alpha=0.5)
ax.axvline(x=y_mean, color='gray', linestyle='--', alpha=0.5)

# 设置标签和标题
ax.set_xlabel('跨境数据传输量', fontsize=12)
ax.set_ylabel('空间滞后（邻居平均值）', fontsize=12)
ax.set_title(f'跨境数据传输量Moran散点图 (2023)\nMoran\'s I = {moran_i:.4f}, p = {p_value:.4f}', 
             fontsize=14, fontweight='bold')

# 添加图例和网格
ax.legend(loc='best')
ax.grid(True, alpha=0.3)

# 添加颜色条
cbar = plt.colorbar(scatter)
cbar.set_label('跨境数据传输量', fontsize=10)

plt.tight_layout()
plt.savefig('../outputs/figures/global_moran.png', dpi=300, bbox_inches='tight')
plt.show()

# 保存Moran's I结果
moran_results = pd.DataFrame({
    '指标': ['跨境数据传输总量_TB'],
    'Moran_I': [moran_i],
    '期望值': [expected_i],
    '方差': [var_i],
    'Z得分': [z_score],
    'P值': [p_value],
    '显著性(α=0.05)': [p_value < 0.05],
    '城市数量': [n]
})
moran_results.to_csv('../outputs/tables/global_moran.csv', index=False, encoding='utf-8-sig')
# %% [markdown]
# # 5. 高级统计分析

# %%
# 5.1 基尼系数计算（城市间不平等）
def gini_coefficient(x):
    """计算基尼系数"""
    x = np.array(x)
    x = x.flatten()
    if np.amin(x) < 0:
        x -= np.amin(x)
    x = np.sort(x)
    n = len(x)
    cumx = np.cumsum(x)
    return (n + 1 - 2 * np.sum(cumx) / cumx[-1]) / n

# 计算各年度基尼系数
gini_results = []
for year in df_processed['年份'].unique():
    year_data = df_processed[df_processed['年份'] == year]
    gini = gini_coefficient(year_data['跨境数据传输总量_TB'].fillna(0).values)
    gini_results.append({'年份': year, '基尼系数': gini})

gini_df = pd.DataFrame(gini_results)
print("=== 跨境数据传输量基尼系数 ===")
print(gini_df.round(4))

# 绘制基尼系数变化趋势
plt.figure(figsize=(10, 6))
plt.plot(gini_df['年份'], gini_df['基尼系数'], marker='o', linewidth=2, markersize=8)
plt.fill_between(gini_df['年份'], gini_df['基尼系数'], alpha=0.2)
plt.title('跨境数据传输量基尼系数变化趋势', fontsize=14, fontweight='bold')
plt.xlabel('年份')
plt.ylabel('基尼系数')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/figures/gini_coefficient.png', dpi=300, bbox_inches='tight')
plt.show()

# %%
# 5.2 泰尔指数分解
def theil_index(x):
    """计算泰尔指数"""
    x = np.array(x)
    mean_x = np.mean(x)
    n = len(x)
    # 避免除以0和log(0)
    x = np.where(x <= 0, 0.0001, x)
    return np.sum((x / mean_x) * np.log(x / mean_x)) / n

# 计算泰尔指数
theil_results = []
for year in df_processed['年份'].unique():
    year_data = df_processed[df_processed['年份'] == year]
    theil = theil_index(year_data['跨境数据传输总量_TB'].fillna(0.0001).values)
    theil_results.append({'年份': year, '泰尔指数': theil})

theil_df = pd.DataFrame(theil_results)
print("=== 跨境数据传输量泰尔指数 ===")
print(theil_df.round(4))

# 保存基尼系数和泰尔指数结果
inequality_df = pd.merge(gini_df, theil_df, on='年份')
inequality_df.to_csv('../outputs/tables/inequality_indices.csv', index=False, encoding='utf-8-sig')

# %%
# 5.3 洛伦兹曲线绘制
def lorenz_curve(x):
    """计算洛伦兹曲线坐标"""
    x = np.sort(x)
    n = len(x)
    cumx = np.cumsum(x)
    cumulative_wealth = cumx / cumx[-1]
    cumulative_population = np.arange(1, n + 1) / n
    return cumulative_population, cumulative_wealth

# 绘制2023年洛伦兹曲线
data_2023 = df_processed[df_processed['年份'] == 2023]
transmission_data = data_2023['跨境数据传输总量_TB'].fillna(0).values

cum_pop, cum_wealth = lorenz_curve(transmission_data)

plt.figure(figsize=(10, 8))
plt.plot(cum_pop, cum_wealth, label='洛伦兹曲线', linewidth=3)
plt.plot([0, 1], [0, 1], 'k--', label='绝对平等线', alpha=0.5)
plt.fill_between(cum_pop, cum_pop, cum_wealth, alpha=0.2, color='red', label='不平等区域')

plt.title('跨境数据传输量洛伦兹曲线 (2023)', fontsize=14, fontweight='bold')
plt.xlabel('城市累积比例', fontsize=12)
plt.ylabel('数据传输量累积比例', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('../outputs/figures/lorenz_curve.png', dpi=300, bbox_inches='tight')
plt.show()

# %% [markdown]
# # 6. OD矩阵分析

# %%
# 6.1 网络流量分析
# 计算各年度总流量
total_flow = {}
for year, od_df in od_data.items():
    total_flow[year] = od_df.values.sum()

total_flow_df = pd.DataFrame(list(total_flow.items()), columns=['年份', '总流量'])
print("=== OD矩阵总流量 ===")
print(total_flow_df)

# 绘制流量变化趋势
plt.figure(figsize=(10, 6))
plt.bar(total_flow_df['年份'], total_flow_df['总流量'], color='steelblue')
plt.title('大湾区城市间数据传输总流量变化', fontsize=14, fontweight='bold')
plt.xlabel('年份')
plt.ylabel('总流量(TB)')
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.savefig('../outputs/figures/od_total_flow.png', dpi=300, bbox_inches='tight')
plt.show()

# %%
# 6.2 城市中心性分析
def calculate_centrality(od_matrix):
    """计算网络中心性指标"""
    # 度中心性
    in_degree = od_matrix.sum(axis=0)  # 入度
    out_degree = od_matrix.sum(axis=1)  # 出度
    total_degree = in_degree + out_degree
    
    # 特征向量中心性（简化版本，使用幂迭代法）
    try:
        # 使用幂迭代法计算特征向量中心性
        n = len(od_matrix)
        eigenvector = np.ones(n) / n
        
        for _ in range(100):
            new_eigenvector = od_matrix @ eigenvector
            new_eigenvector = new_eigenvector / np.linalg.norm(new_eigenvector)
            
            if np.linalg.norm(new_eigenvector - eigenvector) < 1e-6:
                break
            eigenvector = new_eigenvector
        
        eigenvector_centrality = np.abs(eigenvector)
    except:
        eigenvector_centrality = np.ones(len(od_matrix))
    
    return {
        '入度': in_degree,
        '出度': out_degree,
        '总度': total_degree,
        '特征向量中心性': eigenvector_centrality
    }

# 计算2023年中心性
od_2023 = od_data[2023].values
centrality_2023 = calculate_centrality(od_2023)

# 创建中心性DataFrame
cities = od_data[2023].index
centrality_df = pd.DataFrame({
    '城市': cities,
    '入度中心性': centrality_2023['入度'],
    '出度中心性': centrality_2023['出度'],
    '总度中心性': centrality_2023['总度'],
    '特征向量中心性': centrality_2023['特征向量中心性']
})

print("=== 2023年城市中心性排名 ===")
print(centrality_df.sort_values('总度中心性', ascending=False).head())

# 绘制中心性雷达图
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

top_cities = centrality_df.nlargest(6, '总度中心性')['城市']

for idx, city in enumerate(top_cities):
    ax = axes[idx]
    values = [
        centrality_df.loc[centrality_df['城市'] == city, '入度中心性'].values[0],
        centrality_df.loc[centrality_df['城市'] == city, '出度中心性'].values[0],
        centrality_df.loc[centrality_df['城市'] == city, '总度中心性'].values[0],
        centrality_df.loc[centrality_df['城市'] == city, '特征向量中心性'].values[0] * 1000  # 缩放
    ]
    
    angles = np.linspace(0, 2 * np.pi, len(values), endpoint=False)
    values = np.concatenate((values, [values[0]]))
    angles = np.concatenate((angles, [angles[0]]))
    
    ax.plot(angles, values, 'o-', linewidth=2)
    ax.fill(angles, values, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(['入度', '出度', '总度', '特征向量'])
    ax.set_title(f'{city}中心性分析', fontsize=12, fontweight='bold')
    ax.grid(True)

plt.suptitle('主要城市网络中心性雷达图 (2023)', fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.savefig('../outputs/figures/centrality_radar.png', dpi=300, bbox_inches='tight')
plt.show()

# 保存中心性分析结果
centrality_df.to_csv('../outputs/tables/network_centrality.csv', index=False, encoding='utf-8-sig')

# %% [markdown]
# # 7. 数据质量评估报告

# %%
# 7.1 数据完整性评估
completeness_report = pd.DataFrame({
    '指标': key_metrics,
    '总样本数': [len(df_processed)] * len(key_metrics),
    '非空样本数': [df_processed[metric].notnull().sum() for metric in key_metrics],
    '缺失样本数': [df_processed[metric].isnull().sum() for metric in key_metrics],
    '缺失率(%)': [(df_processed[metric].isnull().sum() / len(df_processed)) * 100 for metric in key_metrics]
})

print("=== 数据完整性评估报告 ===")
print(completeness_report.round(2))

# 保存数据质量报告
completeness_report.to_csv('../outputs/tables/data_quality_report.csv', index=False, encoding='utf-8-sig')

# %%
# 7.2 异常值检测（不使用scikit-learn，改用统计方法）
print("=== 异常值检测报告（使用IQR方法） ===")

outlier_report = []
for metric in key_metrics[:6]:  # 只检查前6个关键指标
    data = df_processed[metric].dropna().values
    
    if len(data) > 0:
        # 使用IQR方法检测异常值
        Q1 = np.percentile(data, 25)
        Q3 = np.percentile(data, 75)
        IQR = Q3 - Q1
        
        # 定义异常值边界
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        # 计算异常值
        outliers = data[(data < lower_bound) | (data > upper_bound)]
        outlier_count = len(outliers)
        outlier_percent = (outlier_count / len(data)) * 100
        
        # 使用Z-score方法作为补充（|Z| > 3视为异常）
        z_scores = np.abs((data - np.mean(data)) / np.std(data))
        z_outliers = data[z_scores > 3]
        z_outlier_count = len(z_outliers)
        z_outlier_percent = (z_outlier_count / len(data)) * 100
        
        outlier_report.append({
            '指标': metric,
            '总样本数': len(data),
            'IQR异常值数量': outlier_count,
            'IQR异常值比例(%)': outlier_percent,
            'Z-score异常值数量': z_outlier_count,
            'Z-score异常值比例(%)': z_outlier_percent
        })
        
        # 可视化异常值
        if outlier_count > 0 or z_outlier_count > 0:
            fig, axes = plt.subplots(1, 3, figsize=(15, 4))
            
            # 1. 箱线图
            axes[0].boxplot(data)
            axes[0].set_title(f'{metric}箱线图', fontsize=12)
            axes[0].set_ylabel('值')
            axes[0].grid(True, alpha=0.3)
            
            # 2. 直方图
            axes[1].hist(data, bins=30, alpha=0.7, color='steelblue', edgecolor='black')
            axes[1].axvline(x=lower_bound, color='red', linestyle='--', label=f'下界: {lower_bound:.2f}')
            axes[1].axvline(x=upper_bound, color='red', linestyle='--', label=f'上界: {upper_bound:.2f}')
            axes[1].axvline(x=np.mean(data), color='green', linestyle='-', label=f'均值: {np.mean(data):.2f}')
            axes[1].set_title(f'{metric}分布', fontsize=12)
            axes[1].set_xlabel('值')
            axes[1].set_ylabel('频数')
            axes[1].legend(fontsize=8)
            axes[1].grid(True, alpha=0.3)
            
            # 3. Z-score散点图
            axes[2].scatter(range(len(z_scores)), z_scores, alpha=0.6, color='orange')
            axes[2].axhline(y=3, color='red', linestyle='--', label='Z=3（异常值边界）')
            axes[2].axhline(y=-3, color='red', linestyle='--')
            axes[2].set_title(f'{metric} Z-score', fontsize=12)
            axes[2].set_xlabel('样本索引')
            axes[2].set_ylabel('Z-score')
            axes[2].legend()
            axes[2].grid(True, alpha=0.3)
            
            plt.suptitle(f'{metric}异常值检测', fontsize=14, fontweight='bold', y=1.05)
            plt.tight_layout()
            plt.savefig(f'../outputs/figures/outliers_{metric}.png', dpi=300, bbox_inches='tight')
            plt.show()

outlier_df = pd.DataFrame(outlier_report)
print(outlier_df.round(2))

# 保存异常值报告
outlier_df.to_csv('../outputs/tables/outlier_report.csv', index=False, encoding='utf-8-sig')
# %% [markdown]
# # 8. 总结与发现

# %%
# 生成探索性数据分析总结
summary = {
    '数据基本情况': {
        '数据时间跨度': f"{df_main['年份'].min()}年 - {df_main['年份'].max()}年",
        '城市数量': df_main['城市'].nunique(),
        '总观测值': len(df_main),
        '指标数量': len(df_main.columns) - 3  # 减去城市、年份、城市代码
    },
    '数据质量': {
        '平均缺失率': f"{completeness_report['缺失率(%)'].mean():.2f}%",
        '关键指标完整性': f"{(completeness_report['缺失率(%)'] < 5).sum()}/{len(key_metrics)}个指标缺失率<5%"
    },
    '统计特征': {
        '基尼系数范围': f"{gini_df['基尼系数'].min():.3f} - {gini_df['基尼系数'].max():.3f}",
        '泰尔指数趋势': '上升' if theil_df['泰尔指数'].iloc[-1] > theil_df['泰尔指数'].iloc[0] else '下降',
        '空间自相关': '需要安装esda库' if knn is None else '显著' if 'moran' in locals() and moran.p_norm < 0.05 else '不显著'
    },
    '网络特征': {
        '总流量增长': f"{((total_flow_df.iloc[-1]['总流量'] - total_flow_df.iloc[0]['总流量']) / total_flow_df.iloc[0]['总流量'] * 100):.1f}%" if len(total_flow_df) > 0 else 'N/A',
        '中心性最高城市': centrality_df.loc[centrality_df['总度中心性'].idxmax(), '城市'] if len(centrality_df) > 0 else 'N/A'
    }
}

print("=== 探索性数据分析总结 ===")
for category, items in summary.items():
    print(f"\n{category}:")
    for key, value in items.items():
        print(f"  {key}: {value}")

# 保存分析总结
summary_df = pd.DataFrame([
    {'类别': cat, '指标': k, '结果': v} 
    for cat, items in summary.items() 
    for k, v in items.items()
])
summary_df.to_csv('../outputs/tables/eda_summary.csv', index=False, encoding='utf-8-sig')

print("\n" + "="*60)
print("✅ 探索性数据分析完成！")
print("="*60)
print(f"📊 图表保存至: ../outputs/figures/")
print(f"📋 表格保存至: ../outputs/tables/")
print(f"📈 主要发现:")
print(f"  1. 数据时间跨度: {df_main['年份'].min()}年 - {df_main['年份'].max()}年")
print(f"  2. 分析城市数量: {df_main['城市'].nunique()}个")
print(f"  3. 总流量增长: {((total_flow_df.iloc[-1]['总流量'] - total_flow_df.iloc[0]['总流量']) / total_flow_df.iloc[0]['总流量'] * 100):.1f}%" if len(total_flow_df) > 0 else 'N/A')
print("="*60)
```
